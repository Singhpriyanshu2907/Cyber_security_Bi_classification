```python
## Below function will be used to drop Variables with std_dev 0 as those variable wont contribute much in predictions
def get_cols_with_zero_std_dev(df: pd.DataFrame):
    """
    Returns a list of columns names which are having zero standard deviation.
    """
    cols_to_drop = []
    num_cols = [col for col in df.columns if df[col].dtype != 'O']  # numerical cols only
    for col in num_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return df.drop(columns=cols_to_drop)


## Function for Outlier treatment
def treat_outliers(df, lower_quantile=0.01, upper_quantile=0.99):

## This function will return the dataset after clipping at lower percentile of 1% & upper percentile of 99% by iterating over each column
    treated_df = pd.DataFrame()

    for column in df.columns:
        lower_threshold = df[column].quantile(lower_quantile)
        upper_threshold = df[column].quantile(upper_quantile)

        treated_column = df[column].clip(lower_threshold, upper_threshold)
        treated_df[column] = treated_column

    return treated_df
```

```python
## Dropping the variables with 0 std_dev
X = get_cols_with_zero_std_dev(X)
```

```python
## Creating a correlation plot to check multicolinearity
plt.figure(figsize=(25,20))
sns.heatmap(data=X.corr(),annot=True, cmap=plt.cm.CMRmap_r)
plt.show()

"""

From the below Heatmap it can be clearly seen that there are multiple columns which are highly colinear with other columns,
So need to drop those coulmns to get rid off of the issue of multicolinearity.

"""
```

```python
# Calculate the correlation matrix
corr_matrix = X.corr().abs()

# Create a boolean mask to identify columns with correlation greater than 0.90
high_corr_cols = np.full(corr_matrix.shape, False, dtype=bool)
high_corr_cols[np.triu_indices(len(high_corr_cols), k=1)] = corr_matrix.values[np.triu_indices(len(corr_matrix), k=1)] > 0.90

# Get the column indices to drop
cols_to_drop = np.where(high_corr_cols.any(axis=0))[0]

# Drop the columns from the DataFrame
X.drop(X.columns[cols_to_drop], axis=1, inplace=True)
```

```python
## Checking for outliers by Boxplot
plt.figure(figsize=(15,10))
sns.boxplot(data=X,orient="h")
plt.title("Dataset before Outlier treatment")
plt.show()

## Treating the outliers
X = treat_outliers(X)

## Checking for outliers by Boxplot
plt.figure(figsize=(15,10))
sns.boxplot(data=X,orient="h")
plt.title("Dataset after Outlier treatment")
plt.show()
```

```python
## Splitting the dataset into train & test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(f"train set: {X_train.shape, y_train.shape}")
print(f"test set: {X_test.shape, y_test.shape}")
```

```python
## As we can see it is a imbalanced dataset
y_train.value_counts()
```

```python
## Initializing resampling technique SMOTE under resampler
resampler = SMOTE(random_state = 42)
X_train_res, y_train_res = resampler.fit_resample(X_train,y_train)
```

```python
# Dataset is balaned now
y_train_res.value_counts()
```

```python
# Define the column names
numerical_cols = list(X_train_res.columns)


# Create the preprocessing steps for numerical columns
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), ## imputing missing values if found any
    ('scaler', StandardScaler()) ## scalung up the data using standard scaler
])



# Create the column transformer
preprocessor_X = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical_cols)
])



# Apply the preprocessing pipeline to Train_X and Test_X
X_train_sc = pd.DataFrame(preprocessor_X.fit_transform(X_train_res),columns=preprocessor_X.get_feature_names_out())
X_test_sc = pd.DataFrame(preprocessor_X.transform(X_test),columns=preprocessor_X.get_feature_names_out())
```

```python
# Define the column transformer
preprocessor_Y = ColumnTransformer(
    transformers=[
        ('ordinalencoder',OrdinalEncoder(categories=[["No","Yes"]]),["Attack"]) ## Applying encoding on target variable
    ]
)

# Apply the preprocessing pipeline to Y_tarin & Y_test, Converting "No" as 0 & "Yes" as 1
y_train_sc = pd.DataFrame(preprocessor_Y.fit_transform(y_train_res), columns=preprocessor_Y.get_feature_names_out(["Attack"]))
y_test_sc = pd.DataFrame(preprocessor_Y.fit_transform(y_test), columns=preprocessor_Y.get_feature_names_out(["Attack"]))
```

#### Creating a model before Feautre selection

##### Created 2 Models which are Decision Tree & Logistic regression & both the models resulted in 100% accuracy, so there might be a issue of overfitting.

```python
## Initializing Decision Tree model
DT_class = DecisionTreeClassifier()

## Training the model
DT_class.fit(X_train_sc,y_train_sc)

## Predicting for Train & Test dataset
train_pred_DT = DT_class.predict(X_train_sc)
test_pred_DT = DT_class.predict(X_test_sc)

## Printing classification report for Train & Test
print(classification_report(y_train_sc, train_pred_DT))
print(classification_report(y_test_sc, test_pred_DT))

## Calculating ROC_AUC Score
train_auc = roc_auc_score(y_train_sc['ordinalencoder__Attack'],train_pred_DT)
test_auc = roc_auc_score(y_test_sc['ordinalencoder__Attack'], test_pred_DT)

## Printing ROC_AUC Score for Train & Test
print("The AUC for the model built on the Train Data is : ", train_auc)
print("The AUC for the model built on the Test Data is : ", test_auc)
```

```python
## Initializing Logistic Regression model
LG_class = LogisticRegression()

## Training the model
LG_class.fit(X_train_sc,y_train_sc)

## Predicting for Train & Test dataset
train_pred_LG= LG_class.predict(X_train_sc)
test_pred_LG = LG_class.predict(X_test_sc)

## Printing classification report for Train & Test
print(classification_report(y_train_sc, train_pred_LG))
print(classification_report(y_test_sc, test_pred_LG))

## Calculating ROC_AUC Score
train_auc = roc_auc_score(y_train_sc['ordinalencoder__Attack'],train_pred_LG)
test_auc = roc_auc_score(y_test_sc['ordinalencoder__Attack'], test_pred_LG)

## Printing ROC_AUC Score for Train & Test
print("The AUC for the model built on the Train Data is : ", train_auc)
print("The AUC for the model built on the Test Data is : ", test_auc)
```

#### Feature Selection

```python

# Create a Random Forest classifier
rf = RandomForestClassifier()

# Fit the classifier to your data
rf.fit(X_train_sc, y_train_sc)

# Get the feature importances from the trained model
feature_importances = rf.feature_importances_

# Get the indices of the features sorted by their importance scores
sorted_indices = feature_importances.argsort()[::-1]

# Get the names of the features in descending order of importance
top_20_feature_names = X_train_sc.columns[sorted_indices[:20]]

# Print the feature names and their corresponding importance scores
for feature_name, importance_score in zip(top_20_feature_names, feature_importances[sorted_indices[:20]]):
    print(feature_name, ": ", importance_score)
```

```python
top_20 = list (top_20_feature_names)
```

```python
## Finding Top 20 important features using RFE
model = LogisticRegression()

rfe = RFE(estimator=model, n_features_to_select=20)

rfe = rfe.fit(X_train_sc, y_train_sc)

imp_vars_RFE = list(X_train_sc.columns[rfe.support_])
```

```python
## Finding Top 20 important features using SKB
SKB = SelectKBest(f_classif, k=20).fit(X_train_sc, y_train_sc)

SKB.get_support()

imp_vars_SKB = list(X_train_sc.columns[SKB.get_support()])
```

```python
#Creating a combined list of the top features by intersecting the result of RFC, RFE, SKB
RFE_features = set(imp_vars_RFE)

SKB_features = set(imp_vars_SKB)

RFC_features = set(top_20)

Final_Var = RFE_features.intersection(SKB_features, RFC_features)

## Selecting the features in train & test dadaset
X_train_FS = X_train_sc[Final_Var]
X_test_FS = X_test_sc[Final_Var]
```

```python
## Creating a list of models
models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    XGBClassifier(),
    SVC(),
    RandomForestClassifier(random_state=42)
]


# Perform cross-validation and get the scores for each model
for model in models:
    scores = cross_val_score(model, X_train_FS, y_train_sc, cv=5)
    print("Model:", type(model).__name__)
    print("Scores:", scores)
    print("Average Score:", scores.mean())
    print()
```

```python
## Initializing Random Forest model
RF_class = RandomForestClassifier(random_state=42)

## Training the model
RF_class.fit(X_train_FS,y_train_sc)

## Predicting for Train & Test dataset
train_pred_RF = RF_class.predict(X_train_FS)
test_pred_RF = RF_class.predict(X_test_FS)

## Printing classification report for Train & Test
print(classification_report(y_train_sc, train_pred_RF))
print(classification_report(y_test_sc, test_pred_RF))

## Calculating ROC_AUC Score
train_auc = roc_auc_score(y_train_sc['ordinalencoder__Attack'],train_pred_RF)
test_auc = roc_auc_score(y_test_sc['ordinalencoder__Attack'], test_pred_RF)

## Printing ROC_AUC Score for Train & Test
print("The AUC for the model built on the Train Data is : ", train_auc)
print("The AUC for the model built on the Test Data is : ", test_auc)
```
